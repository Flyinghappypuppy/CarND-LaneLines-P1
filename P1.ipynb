Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
class SmoothDraw(object):         
    def __init__(self):
        self.mean_slopes = []
        self.mean_intercepts = []
        self.mean_of_mean_slopes = 0
        self.mean_of_mean_intercepts = 0
        self.smoothing_window = 10
        self.sd_window = 12
        self.sd_fac = 7.5
        
    def draw(self, slopes, intercepts, color, thickness, img):
        
        avg_slope, avg_intercept = self._find_average(slopes, intercepts)
        # When no slope in frame just use previous mean
        if (not avg_slope) or (not avg_intercept) or np.isnan(avg_slope) or np.isnan(avg_intercept):
            # print (avg_slope, avg_intercept)
            pass
        else:
            # Continually calculate mean within smoothing window
            self._update_mean(avg_slope, avg_intercept)

#             y1 = img.shape[0]
#             x1 = np.int((y1 - avg_intercept) / avg_slope)
#             y2 = np.int(img.shape[0] * 0.6)
#             x2 = np.int((y2 - avg_intercept) / avg_slope)
#             cv2.line(img, (x1, y1), (x2, y2), color , thickness)
        
        # Find point and Draw
        y1 = img.shape[0]
        x1 = np.int((y1 - self.mean_of_mean_intercepts) / self.mean_of_mean_slopes)
        y2 = np.int(img.shape[0] * 0.65)
        x2 = np.int((y2 - self.mean_of_mean_intercepts) / self.mean_of_mean_slopes)
        cv2.line(img, (x1, y1), (x2, y2), color , thickness)
    
    # Mean of last two frames
    def _update_mean(self, avg_slope, avg_intercept):
        self.mean_slopes.append(avg_slope)
        self.mean_intercepts.append(avg_intercept)
        
        # Greedily selecting upto self.smoothing_window number of previous means
        ind = self._greedy_last_N(self.mean_slopes, self.smoothing_window)
        
        # Compute Mean
        self.mean_of_mean_slopes =  np.nanmean(np.array(self.mean_slopes[ind:]))
        self.mean_of_mean_intercepts = np.nanmean(np.array(self.mean_intercepts[ind:]))
        
    
    def _find_average(self, slopes, intercepts):
        '''
        Find the average mean and interceprt for the given frame
        '''
        if (not slopes) or (not intercepts): 
            return None,None
        filt_slopes, filt_intercepts = self._filter_outliers(np.array(slopes), np.array(intercepts), self.sd_fac)
        return np.nanmean(filt_slopes), np.nanmean(filt_intercepts)
        
    def _filter_outliers(self, slopes, intercepts, sd_fac):
        '''
        Filter list based on Progressive Standard Deviation
            Input: Numpy array of Slopes and Intercepts, SD factor 
            Output: Return Filtered list
        '''
        if len(self.mean_slopes) <= 5:
            sd_slopes, sd_intercepts = slopes, intercepts
            # print ("If", sd_slopes)
        else:
            # print ("else", self.mean_slopes)
            sd_slopes, sd_intercepts = self.mean_slopes , self.mean_intercepts

            # Greedily selecting upto self.SD_window number of previous means
            ind = self._greedy_last_N(sd_slopes, self.sd_window)
            # print(ind)
            sd_slopes, sd_intercepts = sd_slopes[ind:] , sd_intercepts[ind:]
        # print (sd_slopes)
        s_mean = np.nanmean(sd_slopes)
        s_sd = np.std(sd_slopes)
        i_mean = np.nanmean(sd_intercepts)
        i_sd = np.std(sd_intercepts)
        # print (s_mean, s_sd, i_mean, i_sd)
        non_outliers = ((abs(slopes - s_mean) < sd_fac * s_sd) &
                        (abs(intercepts - i_mean) < sd_fac * i_sd ))
        # print (non_outliers, slopes)
        slopes, intercepts = slopes[non_outliers], intercepts[non_outliers]
        if not len(slopes):
            return self.mean_slopes[-1], self.mean_intercepts[-1]
        else:
            return slopes, intercepts

    def _greedy_last_N(self, values, N):
        '''
        Greedily find upto N last values in the list
            Input: Values list, Size
            Output: Index of max last N
        '''
        if len(values) >= N:
            ind = N * -1
        else:
            ind = (len(values) % N) * -1
        return ind
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
    
def draw_lines_test(img, lines, color, thickness):
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    left_slopes = []
    left_inters = []
    right_slopes = []
    right_inters = []
    left_lines = []
    right_lines = []
    # Parameters
    angle_min = 25
    angle_max = 65
    r_inter_max = 60
    r_inter_min = -20
    l_inter_max = 900
    l_inter_min = 600
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            mb = np.polyfit([x1, x2], [y1, y2], 1)
            slope = mb[0]
            intercept = mb[1]
            # print (slope)
            if (slope > math.tan(math.radians(90 + angle_min)) 
                and slope < math.tan(math.radians(90 + angle_max))
                and intercept > l_inter_min
                and intercept < l_inter_max):
                left_slopes.append(slope)
                left_inters.append(intercept)
                left_lines.append(line)
                # cv2.line(img, (x1, y1), (x2, y2), [255,255,0], 10)
            elif (slope > math.tan(math.radians(angle_min)) 
                  and slope < math.tan(math.radians(angle_max))
                  and intercept > r_inter_min
                  and intercept < r_inter_max):
                right_slopes.append(slope)
                right_inters.append(intercept)
                right_lines.append(line)
                # cv2.line(img, (x1, y1), (x2, y2), [255,255,0], 10)

    ## Render Left Lane of frame
#     draw_lines_test(img, left_lines, [0,0,255], thickness)
    # When slope exists
    if left_slopes and left_inters:
        LeftDraw.draw(left_slopes, left_inters, color, thickness, img)
    # When no useful slope in frame
    else: 
        LeftDraw.draw(None, None, color, thickness, img)
        
    # Render Right Lane of frame
#     draw_lines_test(img, left_lines, [0,0,255], thickness)
    # When slope exists
    if right_slopes and right_inters:
        RightDraw.draw(right_slopes, right_inters, color, thickness, img)
    # When no useful slope in frame
    else: 
        RightDraw.draw(None, None, color, thickness, img)
    
    # print (avg_leftslope, avg_rightslope)
    # print (avg_left_inter, avg_right_inter)
    
def filter_outliers(means, intercepts, bound):
    non_outliers = ((abs(means - np.nanmean(means)) < bound * np.std(means)) &
                    (abs(intercepts - np.nanmean(intercepts)) < bound * np.std(intercepts)))
    return means[non_outliers], intercepts[non_outliers]

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

imagetest = os.listdir("test_images/")

for image_name in imagetest:
    
    #Instantiate Smooth Draw Class
    LeftDraw = SmoothDraw()
    RightDraw = SmoothDraw()
    
    #Get Image
    image = mpimg.imread('test_images/' + image_name)

    # Grayscale the image
    gray = grayscale(image)

    # Apply Gaussian smoothing
    kernel_size = 7
    blur_gray = gaussian_blur(gray, kernel_size)

    # Apply canny
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)
 
    # Select region
    imshape = image.shape
    x_middle = imshape[1]/2
    x_bottomscope = image.shape[1] * 0.45
    x_topscope = 100
    y_max = imshape[0]
    y_viewmax = imshape[0] * 0.65
    vertices = np.array([[(x_middle - x_bottomscope , y_max),
                          (x_middle - x_topscope, y_viewmax),
                          (x_middle + x_topscope , y_viewmax),
                          (x_middle + x_bottomscope, y_max)]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 0.5
    theta = np.pi/360 
    threshold = 15
    min_line_length = 33
    max_line_gap = 35

    # Copy of original image
    orig_image = np.copy(image)

    # Hough on edge detected image
    hough_image = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

    # Combine Images
    result = weighted_img(hough_image, orig_image,1,1)

    plt.imshow(result)
    plt.savefig("test_images_output/" + image_name + ".jpg")
````
